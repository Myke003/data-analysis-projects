```python
import pandas as pd
import openpyxl
```

```python
df = pd.read_excel('data/Base víctimas_31_enero_2025.xlsx', sheet_name='Hoja1', engine='openpyxl')
```

# Data Cleaning

```python
df_selected = df[['departamento', 'municipio','Fecha del accidente', 'Año', 'Mes', 'latitude', 'longitude', 'cod_dane_muni', 'cod_dane_depto','Tipo de lugar', 'Presunto actor responsable','Tipo de área', 'id_imsma evento', 'condicion', 'genero', 'estado','rango de edad', 'actividad', 'id_imsma_victima']].copy()

df_selected.columns = df_selected.columns.str.capitalize()
```

```python
df_selected = df_selected.copy()
df_selected[['Departamento', 'Municipio']] = df_selected[['Departamento', 'Municipio']].apply(lambda x: x.str.capitalize())
```

```python
def limpiar_actor(actor):
  actor = actor.strip().upper()
  
  if actor in ['DESCONOCIDO', 'SIN INFORMACION']:
    return "Desconocido"
  
  if any(sigla in actor for sigla in ["ELN", "FARC", "GAO"]):
    return actor
  
  return actor.title()

df_selected["Presunto actor responsable"] = df_selected["Presunto actor responsable"].apply(limpiar_actor)
```

```python
df_selected["Tipo de lugar"].unique()
```

```python
reemplazos = {
    "Sin informacion": "Sin información",
    "En cerca a edificio residencial": "Cerca de un edificio residencial",
    "en sobre cerca rivera de un rio": "Cerca de la ribera de un río",
    "Canos": "Caños",
    "Infra torres electricas": "Infraestructura de torres eléctricas",
    "Infra gaseoducto": "Infraestructura de gaseoducto",
    "Centros deportivos recreacion": "Centros deportivos y de recreación",
    "En cerca instalacion militar": "Cerca de una instalación militar",
    "En cerca edificio gubernamental": "Cerca de un edificio gubernamental",
    "Infra hidroelectricas": "Infraestructura hidroeléctrica",
}

df_selected["Tipo de lugar"] = df["Tipo de lugar"].replace(reemplazos)
```

```python
reemplazos_deptos = {
    "Caqueta": "Caquetá",
    "Cordoba": "Córdoba",
    "Norte de santander": "Norte de Santander",
    "Bolivar": "Bolívar",
    "Valle del cauca": "Valle del Cauca",
    "Vaupes": "Vaupés",
    "Choco": "Chocó",
    "La guajira": "La Guajira",
    "Bogota dc": "Bogotá D.C.",
    "Quindio": "Quindío",
    "Boyaca": "Boyacá",
    "Guainia": "Guainía",
    "Atlantico": "Atlántico",
}

df_selected["Departamento"] = df_selected["Departamento"].replace(reemplazos_deptos)
```
