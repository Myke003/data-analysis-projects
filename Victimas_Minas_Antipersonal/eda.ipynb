```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/minas_antipersonal_limpio.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
victimas_totales = df['Id_imsma_victima'].count()
victimas_edad = df['Rango de edad'].value_counts().reset_index()
print(f"Las victimas totales fueron: {victimas_totales:,.2f}")
print("\nDivididas en: \n")
display(victimas_edad)
```

```python
victimas_departamento = df.groupby('Departamento')['Id_imsma_victima'].count().reset_index(name='Count').sort_values(by='Count', ascending=False)
```

```python
victimas_año = df.groupby('Año')['Id_imsma_victima'].count().reset_index(name='Count')
```

```python
actividades_peligrosas = df.groupby('Actividad')['Id_imsma_victima'].count().reset_index(name='Count')
actividades_mas_peligrosas = actividades_peligrosas.sort_values(by='Count', ascending=False).head(10)
```

```python
actores_responsables = df.groupby('Presunto actor responsable')['Id_imsma_victima'].count().reset_index(name='Count')
actores_responsables = actores_responsables.sort_values(by='Count', ascending=False)

actores_responsables
```

```python
estado_victimas = df.groupby('Estado')['Id_imsma_victima'].count().reset_index(name='Count')
estado_victimas
```
